```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
```

```python
def generate_data(K, N):
    # Draw K parameters/coefficients Î¸i (i from 1 to K) from a Gaussian distribution
    theta = np.random.normal(loc=0,scale=1,size=K)
    
    # Generate covariates for N samples
    X = np.random.normal(loc=0, scale=2, size=(N, K))
    
    # Generate output for N samples
    epsilon = np.random.normal(loc=0, scale=(0.1)**(1/2), size=1)
    Y = np.dot(X, theta) + epsilon
    
    return X,theta,Y
```

```python
def infer_theta(X, y, method="OLS", lambda_=0):
    if method == "OLS":
        # Infer theta using OLS
        try:
            theta = np.linalg.inv(X.T @ X) @ X.T @ y
        except np.linalg.LinAlgError:
            print("OLS method failed due to N < K")
            return None
```

```python
def ridge_theta(X,Y,lambda_):
    # Infer theta using Ridge Regression
    ridge = Ridge(alpha=lambda_, fit_intercept=False)
    ridge.fit(X, Y)
    theta = ridge.coef_
    return theta
```

```python
def caculate_l2_distance(thetastar,thetacap):
    return np.linalg.norm(thetastar-thetacap)
```

```python
lambdas = [0.01,0.1,0.2,0.5,1,10,20,50,100,1000]
```

## k = 10

```python
X_costly,theta_costly,Y_costly = generate_data(K=10,N=2)
```

```python
X_cheap,theta_cheap,Y_cheap = generate_data(K=10,N=100)
```

```python
X_not_too_costly,theta_not_too_costly,Y_not_too_costly = generate_data(K=10,N=20)
```

```python
theta_cap_costly = infer_theta(X_costly,Y_costly)
```

```python
theta_cap_cheap = infer_theta(X_cheap,Y_cheap)
```

```python
theta_cap_not_too_costly = infer_theta(X_not_too_costly,Y_not_too_costly)
```

```python
l2_distance_costly = []
for la in lambdas:
    theta_cap = ridge_theta(X_costly,Y_costly,la)
    l2_distance_costly.append(caculate_l2_distance(theta_costly,theta_cap))
```

```python
plt.plot(lambdas,l2_distance_costly)
plt.xlabel('lambdas')
plt.ylabel('L2 distance')
plt.title('N<K Generated DATA L2 Distance');
```

```python
l2_distance_cheap = []
for la in lambdas:
    theta_cap = ridge_theta(X_cheap,Y_cheap,la)
    l2_distance_cheap.append(caculate_l2_distance(theta_cheap,theta_cap))
```

```python
plt.plot(lambdas,l2_distance_cheap)
plt.xlabel('lambdas')
plt.ylabel('L2 distance')
plt.title('N>K Generated DATA L2 Distance');
```

```python
l2_distance_not_too_costly = []
for la in lambdas:
    theta_cap = ridge_theta(X_not_too_costly,Y_not_too_costly,la)
    l2_distance_not_too_costly.append(caculate_l2_distance(theta_not_too_costly,theta_cap))
```

```python
plt.plot(lambdas,l2_distance_not_too_costly)
plt.xlabel('lambdas')
plt.ylabel('L2 distance')
plt.title('N=10K Generated DATA L2 Distance');
```
