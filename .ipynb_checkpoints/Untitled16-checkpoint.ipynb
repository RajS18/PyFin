```python
import numpy as np

# Set parameters
a = 0.5
b = 0.6
r = 0.4
input_size = 2
hidden_size = 10
output_size = 1
learning_rate = 0.1
num_epochs = 1000
num_train_samples = 100
num_test_samples = 100

# Define sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate training data
train_data = np.random.uniform(0, 1, size=(num_train_samples, input_size))
train_labels = np.zeros((num_train_samples, output_size))
for i in range(num_train_samples):
    if (train_data[i][0] - a) ** 2 + (train_data[i][1] - b) ** 2 < r ** 2:
        train_labels[i] = 1

# Generate test data
test_data = np.random.uniform(0, 1, size=(num_test_samples, input_size))
test_labels = np.zeros((num_test_samples, output_size))
for i in range(num_test_samples):
    if (test_data[i][0] - a) ** 2 + (test_data[i][1] - b) ** 2 < r ** 2:
        test_labels[i] = 1

# Initialize weights randomly
W1 = np.random.normal(scale=0.1, size=(input_size, hidden_size))
b1 = np.zeros((1, hidden_size))
W2 = np.random.normal(scale=0.1, size=(hidden_size, output_size))
b2 = np.zeros((1, output_size))

# Train network using error backpropagation
for epoch in range(num_epochs):
    # Forward pass
    hidden = sigmoid(np.dot(train_data, W1) + b1)
    output = sigmoid(np.dot(hidden, W2) + b2)
    
    # Backward pass
    output_error = (train_labels - output) * output * (1 - output)
    hidden_error = output_error.dot(W2.T) * hidden * (1 - hidden)
    
    # Update weights
    W2 += learning_rate * hidden.T.dot(output_error)
    b2 += learning_rate * np.sum(output_error, axis=0, keepdims=True)
    W1 += learning_rate * train_data.T.dot(hidden_error)
    b1 += learning_rate * np.sum(hidden_error, axis=0, keepdims=True)

# Test network
hidden = sigmoid(np.dot(test_data, W1) + b1)
output = sigmoid(np.dot(hidden, W2) + b2)
predictions = np.round(output)
accuracy = np.mean(predictions == test_labels)
print("Accuracy:", accuracy)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the step function
def step(x):
    return np.where(x >= 0, 1, -1)

# Define the circle function
def circle(x, y, a, b, r):
    return np.where((x-a)**2 + (y-b)**2 < r**2, 1, -1)

# Set the circle parameters
a = 0.5
b = 0.6
r = 0.4

# Generate the training and testing data
np.random.seed(0)
train_data = np.random.rand(100, 2)
train_labels = circle(train_data[:,0], train_data[:,1], a, b, r)
test_data = np.random.rand(100, 2)
test_labels = circle(test_data[:,0], test_data[:,1], a, b, r)

# Set the network parameters
input_size = 2
hidden_size = 10
output_size = 1
learning_rate = 0.1
num_epochs = 100
num_runs = 10

# Initialize the weights
weights = [np.random.randn(input_size, hidden_size), np.random.randn(hidden_size, output_size)]

# Train the network
accuracies = []
hyperplanes = []
for run in range(num_runs):
    for epoch in range(num_epochs):
        # Forward pass
        hidden_activations = sigmoid(np.dot(train_data, weights[0]))
        output_activations = sigmoid(np.dot(hidden_activations, weights[1]))
        
        # Backward pass
        output_errors = (train_labels.reshape(-1,1) - output_activations) * output_activations * (1 - output_activations)
        hidden_errors = np.dot(output_errors, weights[1].T) * hidden_activations * (1 - hidden_activations)
        
        # Update weights
        weights[1] += learning_rate * np.dot(hidden_activations.T, output_errors)
        weights[0] += learning_rate * np.dot(train_data.T, hidden_errors)
        
    # Evaluate the accuracy and hyperplanes
    hidden_activations = step(np.dot(train_data, weights[0]))
    output_activations = step(np.dot(hidden_activations, weights[1]))
    accuracy = np.mean(output_activations == train_labels)
    accuracies.append(accuracy)
    hyperplanes.append(weights[0])
    
# Print the accuracies
print("Accuracies:", accuracies)
```

```python
# Plot the hyperplanes
for i in range(hidden_size):
    w = hyperplanes[0][:,i]
    b = 0.5 * (np.max(train_data[:,0]) + np.min(train_data[:,0])) - 0.5 * (np.max(train_data[:,1]) + np.min(train_data[:,1]))
    x = np.linspace(0, 1, 100)
    y = (-w[0] * x - b) / w[1]
    plt.plot(x, y, 'k--')
plt.scatter(train_data[:,0], train_data[:,1], c=train_labels, cmap='bwr')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

class TwoLayerNet:
    def __init__(self, input_size, hidden_size, output_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        
        self.weights1 = np.random.randn(input_size, hidden_size)
        self.weights2 = np.random.randn(hidden_size, output_size)
        
    def forward(self, x):
        self.hidden = sigmoid(np.dot(x, self.weights1))
        self.output = sigmoid(np.dot(self.hidden, self.weights2))
        
    def backward(self, x, y, learning_rate):
        output_error = y - self.output
        output_delta = output_error * sigmoid_derivative(self.output)
        
        hidden_error = np.dot(output_delta, self.weights2.T)
        hidden_delta = hidden_error * sigmoid_derivative(self.hidden)
        
        self.weights2 += np.dot(self.hidden.T, output_delta) * learning_rate
        self.weights1 += np.dot(x.T, hidden_delta) * learning_rate
        
    def train(self, x, y, learning_rate, epochs):
        for i in range(epochs):
            self.forward(x)
            self.backward(x, y, learning_rate)
            
    def predict(self, x):
        self.forward(x)
        return self.output.round()

# generate data
a = 0.5
b = 0.6
r = 0.4
num_samples = 200
X_train = np.random.rand(num_samples, 2)
y_train = np.zeros((num_samples, 1))
for i in range(num_samples):
    if ((X_train[i][0]-a)**2 + (X_train[i][1]-b)**2 < r**2):
        y_train[i] = 1

X_test = np.random.rand(num_samples, 2)
y_test = np.zeros((num_samples, 1))
for i in range(num_samples):
    if ((X_test[i][0]-a)**2 + (X_test[i][1]-b)**2 < r**2):
        y_test[i] = 1

# train the network
input_size = 2
hidden_size = 10
output_size = 1
learning_rate = 0.1
epochs = 1000
nn = TwoLayerNet(input_size, hidden_size, output_size)
nn.train(X_train, y_train, learning_rate, epochs)

# test the network
y_pred = nn.predict(X_test)
accuracy = np.mean(y_pred == y_test)
print("Accuracy:", accuracy)


# plot the hyperplanes corresponding to the hidden nodes
import matplotlib.pyplot as plt
x1 = np.linspace(0, 1, 100)
x2 = np.linspace(0, 1, 100)
xx1, xx2 = np.meshgrid(x1, x2)
X = np.hstack((xx1.reshape(-1,1), xx2.reshape(-1,1)))
hidden_output = sigmoid(np.dot(X, nn.weights1))
for i in range(hidden_size):
    z = hidden_output[:,i].reshape(100,100)
    z = np.where(z > 0.5, 1, 0)
    plt.contour(xx1, xx2, z, levels=[0.5], colors='r')
plt.scatter(X_test[:,0], X_test[:,1], c=y_test[:,0])
plt.show()
```
